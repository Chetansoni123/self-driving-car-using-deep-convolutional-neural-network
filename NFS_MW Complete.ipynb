```python
import tensorflow as tf

from keras.preprocessing.image import load_img, array_to_img, img_to_array, ImageDataGenerator

datagen=ImageDataGenerator(rescale=1./255,
                           height_shift_range=0.2,
                           width_shift_range=0.2,
                           zoom_range=0.2,
                           shear_range=0.2,
                           
                           horizontal_flip=True,
                           rotation_range=40,
                           fill_mode='nearest')
```

```python
# getting model

from tensorflow.keras.applications.vgg16 import VGG16

x= VGG16(include_top = False , weights = 'imagenet', input_shape = (126,55,3))
```

```python
# Defining model and layers arrangement

import tensorflow
# from keras.models import Model
# from keras.layers import Flatten, Dense
from tensorflow.keras.callbacks import ModelCheckpoint,TensorBoard

for layer in x.layers:
    layer.trainable=False
    

y=tensorflow.keras.layers.Flatten()(x.output)
prediction=tensorflow.keras.layers.Dense(3, activation='softmax')(y)

model=tensorflow.keras.Model(inputs=x.input, outputs=prediction)
model.compile(loss='categorical_crossentropy',
              optimizer=tensorflow.keras.optimizers.Adam(learning_rate=0.001),metrics=['accuracy'])
# checkpoint
filepath="weights.{epoch:02d}-{val_loss:.2f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=0, save_best_only=True, mode='max')
```

```python
#create callback list

# Load the TensorBoard notebook extension
%load_ext tensorboard
import datetime,os
# from tensorflow.keras.callbacks import TensorBoard
# NAME = "Self_driving_car{}".format(int(time.time()))
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = TensorBoard(logdir)

```

```python
%reload_ext tensorboard
```

```python
train=datagen.flow_from_directory('Trainingdatagame2', target_size=(126,55),batch_size=32, class_mode='categorical', shuffle = True)
test=datagen.flow_from_directory('Validationdatagame2', target_size=(126,55), batch_size=16, class_mode='categorical', shuffle = True)
```

```python
# from tensorflow.keras.models import load_model

# model = load_model('weights-improvement-02-0.33.hdf5')
model.fit(x=train, validation_data = test,epochs=8, steps_per_epoch=len(train), validation_steps = len(test), callbacks=[tensorboard_callback,checkpoint],verbose=1,sample_weight=None)
```

```python
# code for controlling the game and the output of DCNN/ model

from directkeys import PressKey, ReleaseKey, W,A,S,D
from tensorflow.keras.models import load_model
import cv2
import time
from grabscreen import grab_screen
import numpy as np
from getkeys import key_check


def right():
    PressKey(W)
    PressKey(D)
    ReleaseKey(S)
    
def left():
    PressKey(W)
    PressKey(A)
    ReleaseKey(D)
    
def forward():
    PressKey(W)
    ReleaseKey(A)
    ReleaseKey(D)

model = load_model('6th_good_.h5')




def drive():
       
       
            paused = False
            while True:
                if not paused:
                    screen = grab_screen(region = (100,100+200,730,575))
                    screen = cv2.cvtColor(screen , cv2.COLOR_BGR2RGB)
                    screen = cv2.resize(screen, (55,126))
                    screen4D = np.expand_dims(screen, axis = 0)
                    pred = model.predict(screen4D)
                    print(pred)


                    if pred[0][0] > 0.7:
                        right()

                    elif pred[0][1] > 0.5:
                        forward()

                    elif pred[0][2] > 0.7:
                        left()
                        
                    else: 
                        forward()
                        
                        

                keys = key_check()  
                if 'T' in keys:
                    if paused:
                        paused = False
                        time.sleep(1)
                    else:
                        paused = True
                        ReleaseKey(A)
                        ReleaseKey(W)
                        ReleaseKey(D)
                        time.sleep(1)
    
            
            
            
            
for t in list(range(5))[::-1]:
    print(t)
    time.sleep(1)

drive()    
```
